- barplot: 신뢰구간
- x 범주형, y 수치형 (평균, 편차, 에러)

```python
fig = plt.figure(figsize = (20,6))

area01 = fig.add_subplot(1,3,1)
area01.set_title('sex - survived')
sns.barplot(ax = area01, data = titanic_frm, x = 'sex', y = 'survived')

area02 = fig.add_subplot(1,3,2)
area02.set_title('pclass - fare')
sns.barplot(ax = area02, data = titanic_frm, x = 'sex', y = 'survived', hue = 'class')

area03 = fig.add_subplot(1,3,3)
area02.set_title('pclass - fare')
sns.barplot(ax = area03, data = titanic_frm, x = 'sex', y = 'survived', hue = 'class', dodge = False)

plt.show()
plt.close()
```

```python
# 요일별 팁평균, 요일별 매출평균 - barplot
tips_frm.head()
```

```python
fig = plt.figure(figsize = (20,6))

area01 = fig.add_subplot(1,2,1)
area01.set_title('day - tip')
sns.barplot(ax = area01, data = tips_frm, x = 'day', y = 'tip')

area02 = fig.add_subplot(1,2,2)
area02.set_title('day - total_bill')
sns.barplot(ax = area02, data = tips_frm, x = 'day', y = 'total_bill')

plt.show()
plt.close()
```

- countplot(): 빈도수 표시

```python
fig = plt.figure(figsize = (20,6))

area01 = fig.add_subplot(1,3,1)
area01.set_title('sex - survived')
sns.countplot(ax = area01, data = titanic_frm, x = 'class')

area02 = fig.add_subplot(1,3,2)
area02.set_title('pclass - fare')
sns.countplot(ax = area02, data = titanic_frm, x = 'class', hue = 'who', palette = 'Set3')

area03 = fig.add_subplot(1,3,3)
area02.set_title('pclass - fare')
sns.countplot(ax = area03, data = titanic_frm, x = 'class', hue = 'who', dodge = False)

plt.show()
plt.close()
```

- stripplot(): 모든 데이터를 점으로 시각화

```python
# 요일별로 매출 시각화

plt.figure(figsize = (15,5))

sns.stripplot(data = tips_frm, x = 'day', y = 'total_bill', jitter = True)

plt.show()
plt.close()
```

```python
plt.figure(figsize = (15,5))

sns.swarmplot(data = tips_frm, x = 'day', y = 'total_bill')

plt.show()
plt.close()
```

```python
# 타이타닉 데이터를 서브플롯으로 2개 영역 만들어서 class별 나이 분포 확인 - strip, swarm
fig = plt.figure(figsize = (20,6))

area01 = fig.add_subplot(1,2,1)
area01.set_title('class - age')
sns.stripplot(ax = area01, data = titanic_frm, x = 'class', y = 'age')

area02 = fig.add_subplot(1,2,2)
area02.set_title('class - age')
sns.swarmplot(ax = area02, data = titanic_frm, x = 'class', y = 'age')

plt.show()
plt.close()
```

```python
# 팁 데이터를 서브플롯으로 2개 영역 만들어서 요일, 성별 전체팁의 분포 확인
tips_frm.head()

fig = plt.figure(figsize = (20,6))

area01 = fig.add_subplot(1,2,1)
area01.set_title('strip')
sns.stripplot(ax = area01, data = tips_frm, x = 'day', y = 'tip', hue='sex')

area02 = fig.add_subplot(1,2,2)
area02.set_title('swarm')
sns.swarmplot(ax = area02, data = tips_frm, x = 'day', y = 'tip', hue = 'sex')

plt.show()
plt.close()
```

- boxplot(), violinplot()

```python
fig = plt.figure(figsize = (20,6))

area01 = fig.add_subplot(1,2,1)
area01.set_title('box')
sns.boxplot(ax = area01, data = titanic_frm, x = 'alive', y = 'age')

area02 = fig.add_subplot(1,2,2)
area02.set_title('violin')
sns.violinplot(ax = area02, data = titanic_frm, x = 'alive', y = 'age', hue = 'sex')

plt.show()
plt.close()
```

- heatmap

```python
titanic_frm.head()
```

```python
# 성별에 따른 선실등급의 인원수의 합 - pivot
titanic_frm['cnt'] = 1
table = titanic_frm.pivot_table(index = 'sex', columns = 'class', values = 'cnt', aggfunc = np.sum)
table
```

```python
sns.heatmap(table, annot = True, fmt = 'd', linewidth = 1, cbar = False)
plt.show()
plt.close()
```

```python
# 연도에 월별 승객수에 대한 히트맨 시각화
flights_frm.head()
flight = flights_frm.pivot_table(index = 'month', columns = 'year', values = 'passengers')
sns.heatmap(flight)
```

```python
pop_frm = pd.read_excel('./data/visualization_data/elem_school_folium.xlsx')
pop_frm.head()
```

```python
# Feature Selection
# 학교명, 학교급구분, 소재지도로명주소, 위도, 경도
pop_subset_frm = pop_frm[['학교명','학교급구분','소재지도로명주소','위도','경도']]
# 소재지도로명주소가 서울을 포함하고 학교급구분 초등학교만 추출
result_frm = pop_subset_frm[(pop_subset_frm['학교급구분'] == '초등학교') & (pop_subset_frm['소재지도로명주소'].str.contains('서울특별시'))]
result_frm
```

```python
# 결측값 확인 , 제거
pop_subset_frm.isnull().sum()
pop_subset_frm.dropna(inplace = True)

# 조건에 부합하는 결과를 데이터 프레임으로 만들기

# 지도 시각화
seoul_uni_map = g.Map(location = [37.509671, 127.055517],
                     zoom_start = 12,
                     tiles = 'Stamen Terrain')

for idx in result_frm.index:
    lat = result_frm.loc[idx, '위도']
    lng = result_frm.loc[idx, '경도']
    
    marker01 = g.Marker([lat, lng], 
                      popup = result_frm.loc[idx, '학교명'], 
                      icon = g.Icon(color='blue'))
    marker01.add_to(seoul_uni_map)
    

seoul_uni_map
```
